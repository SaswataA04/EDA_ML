```python
import pandas as pd
```

```python
df = pd.read_csv("encoding_practice_dataset.csv")
```

```python
df
```

```python
# Convert the categorical 'gender' column into dummy/indicator variables (one-hot encoding)
# dtype=int ensures the resulting columns contain integers instead of booleans
# drop_first=True removes the first category(the original column gets deleted) to avoid multicollinearity in modeling
encoded_df = pd.get_dummies(df, columns=['gender'], dtype=int, drop_first=True)
```

```python
encoded_df
#final encoded dataset
```

```python
#here encoding_df_1 means encdoing is being one on already encoded dataframe encoded_df .
# Apply one-hot encoding to the 'marital_status' column in the encoded_df dataframe
# - columns=['marital_status']: Specifies which column to encode in encoded_df
# - dtype=int: Ensures the resulting dummy columns contain integer values (0 or 1)
# - drop_first=True: Removes the first category to avoid multicollinearity 
# The result is stored in a new dataframe called encoded_df_1
encoded_df_1 = pd.get_dummies(encoded_df, columns=['marital_status'], dtype=int, drop_first=True)
```

```python
encoded_df_1
```

```python
# Apply one-hot encoding to multiple categorical columns in encoded_df_1
# - columns=['subscription_plan', 'payment_method', 'device_type']: Specifies which columns to encode
# - dtype=int: Ensures the resulting dummy columns contain integer values (0 or 1)
# - drop_first=True: Removes the first category from each column to avoid multicollinearity
# The result is stored in a new dataframe called encoded_df_2

encoded_df_2 = pd.get_dummies(encoded_df_1, columns=['subscription_plan', 'payment_method', 'device_type'], dtype=int, drop_first=True)

```
